```python
from kiwigrad import draw_dot, Value
```

```python
a = Value(2.)
b = Value(3.)
c = a + b
d = c.log()
e = d.exp()
e.backward()
print(e)
```

```python
draw_dot(e)
```

```python
import torch

at = torch.tensor([2.], requires_grad=True)
bt = torch.tensor([3.], requires_grad=True)
ct = at + bt
dt = ct.log()
et = dt.exp()
et.backward()
print(f"at.grad: {at.grad}, bt.grad: {bt.grad}")
```
